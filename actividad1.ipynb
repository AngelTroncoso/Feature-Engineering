# Transformando Datos: Feature Engineering
## Encoding de Variables Categóricas

¡Hola! Muy bien, hoy vamos a desentrañar uno de los misterios más importantes de los datos: cómo hablarle a una máquina sobre conceptos que no son números. Entender esto cambiará por completo la calidad de tus predicciones y te dará un control total sobre tus modelos.

### Introducción: El lenguaje de los algoritmos

El corazón de un algoritmo es pura aritmética. Para nosotros, conceptos como **México** o **España** son claros, pero para un modelo de aprendizaje automático, son solo cadenas de texto que no puede procesar matemáticamente. Aquí es donde entra el **Encoding**.

El problema real no es solo convertir texto a números, sino cómo hacerlo sin 'mentirle' al modelo. Si asignamos un uno a un país y un dos a otro, el algoritmo podría pensar que el segundo país 'vale el doble', creando un sesgo de orden inexistente.

**El Escenario de Lucía:**
Imagina a Lucía, una analista en una tienda de mascotas. Ella tiene una lista de razas: *Golden Retriever*, *Caniche* y *Bulldog*. Si ella asigna simplemente los números 1, 2 y 3, el modelo podría favorecer a los *Bulldog* solo por tener el número más alto. Lucía se dio cuenta de que el modelo interpretaba una jerarquía donde no la había. Al cambiar su estrategia, logró que cada raza fuera tratada con la misma importancia.

### 1. Label Encoding: Cuando el orden sí importa

Esta técnica es ideal cuando nuestras categorías tienen un orden intrínseco (**ordinales**), como los niveles de educación o el tamaño de una prenda. Aquí, usar números correlativos tiene sentido porque un nivel universitario es, académicamente, superior a un nivel secundario.

En Python, usamos la herramienta `LabelEncoder` de la librería `scikit-learn`. El gran error común es aplicar esto a categorías sin orden (como colores), ya que el modelo intentará buscar promedios matemáticos que no tienen sentido.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Simulamos el dataset de tallas de ropa
data_ropa = {'talla': ['Pequeño', 'Mediano', 'Grande', 'Mediano', 'Pequeño', 'Grande']}
df_ropa = pd.DataFrame(data_ropa)

print('Dataset original:')
print(df_ropa)
print()

# Instanciamos el codificador
encoder = LabelEncoder()

# Aplicamos el encoding a la columna talla
df_ropa['talla_encoded'] = encoder.fit_transform(df_ropa['talla'])

print('Dataset transformado con Label Encoding:')
print(df_ropa)
print()

# Mostramos el mapeo que realizó el algoritmo
mapeo = dict(zip(encoder.classes_, encoder.transform(encoder.classes_)))
print('Mapeo de categorías:')
print(mapeo)
```

### 2. One-Hot Encoding: Igualdad para categorías nominales

¿Qué pasa cuando el orden no existe en absoluto? Aquí brilla el **One-Hot Encoding**. En lugar de una sola columna, creamos columnas nuevas para cada categoría posible. Esta técnica utiliza un sistema binario (unos y ceros) para indicar la presencia de una categoría, eliminando cualquier jerarquía falsa.

Llevaremos esto a la práctica con la función `get_dummies` de `pandas`. Verás cómo pasamos de nombres a una matriz matemática que cualquier red neuronal puede procesar sin inventar superioridades.

```python
# Simulamos datos de ciudades (Variables Nominales)
data_ciudades = {'ciudad': ['Madrid', 'Bogotá', 'Lima', 'Bogotá', 'Madrid']}
df_ciudades = pd.DataFrame(data_ciudades)

print('Dataset original de ciudades:')
print(df_ciudades)
print()

# Aplicamos One-Hot Encoding usando pandas
df_one_hot = pd.get_dummies(df_ciudades, columns=['ciudad'])

print('Dataset con One-Hot Encoding:')
print(df_one_hot)
print()
print('Observa como el valor 1 actúa como un interruptor de encendido para cada ciudad.')
```

### 3. La Trampa de las Variables Ficticias (Dummy Variable Trap)

Al crear tantas columnas podríamos estar repitiendo información innecesaria. Si conocemos el estado de todas las columnas menos una, ya podemos deducir qué hay en la última por descarte. Esto crea **multicolinealidad**, lo que puede desestabilizar algoritmos como la regresión lineal.

La solución técnica es eliminar siempre una de las columnas generadas para mantener la base de datos limpia y eficiente usando el parámetro `drop_first=True`.

```python
# Simulamos datos de género
data_genero = {'genero': ['Masculino', 'Femenino', 'Femenino', 'Masculino']}
df_genero = pd.DataFrame(data_genero)

print('Dataset original de género:')
print(df_genero)
print()

# Aplicamos One-Hot Encoding evitando la trampa de variables ficticias
df_genero_eficiente = pd.get_dummies(df_genero, columns=['genero'], drop_first=True)

print('Dataset optimizado (drop_first=True):')
print(df_genero_eficiente)
print()
print('Si Masculino es 0, por lógica pura el registro es Femenino. ¡Eficiencia total!')
```

### Conclusión y Criterio de Selección

Dominar estas transformaciones es el primer paso real para construir modelos que entiendan el mundo. Recuerda:

*   **Label Encoding:** Solo para variables **ordinales** (con orden lógico).
*   **One-Hot Encoding:** Para variables **nominales** (sin orden).
*   **Pandas vs Scikit-Learn:** Usamos `get_dummies` para análisis rápido y `OneHotEncoder` de Scikit-Learn cuando construimos *Pipelines* de producción.

¿Se dan cuenta de lo fácil que es caer en esa trampa de los números? A veces pensamos que con que el código corra es suficiente, pero la verdadera magia está en entender la naturaleza de nuestra información antes de transformarla.
