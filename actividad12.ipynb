### La Filosofía de la Ingeniería de Características

La ingeniería de características es el arte de traducir la realidad al lenguaje de las máquinas. Un error común es creer que meter más datos siempre es mejor, pero en realidad, la **calidad de la señal** es lo que define el rendimiento. Al aplicar técnicas como el escalado o la codificación, eliminamos los sesgos matemáticos que confunden a los algoritmos, asegurando que cada variable aporte su valor justo sin opacar a las demás.

## El Problema de la Escala y los Datos Faltantes

Imagina un conjunto de datos donde la variable **Edad** tiene huecos y los **Salarios** varían entre cinco mil y cien mil dólares. Sin tratamiento, el modelo ignorará la edad por su escala pequeña frente a los salarios.

A continuación, simularemos este escenario y veremos cómo el uso de `SimpleImputer` y `StandardScaler` normaliza la información para mejorar la precisión del algoritmo.

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

# 1. Creación de datos sintéticos
data = {
    'Edad': [25, 30, np.nan, 45, 50, 22, np.nan, 35],
    'Salario': [5000, 100000, 75000, 40000, 95000, 8000, 60000, 55000]
}
df = pd.DataFrame(data)

print('--- Dataset Original (con nulos y escalas disparas) ---')
print(df)
print()

# 2. Imputación de valores faltantes (Mediana)
imputer = SimpleImputer(strategy='median')
df_imputed = imputer.fit_transform(df)

# 3. Escalado de características (StandardScaler)
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_imputed)

print('--- Dataset Transformado (Valores entre -3 y 3) ---')
print(df_scaled)
print()
print('¡Potente! Ahora la edad y el salario compiten en igualdad de condiciones.')
```

## Automatización: El Escenario de Marina

**Escenario:** Piensa en Marina, una científica de datos en un banco. Ella creó un modelo de crédito que funcionaba perfecto en su computadora, pero al subirlo a la nube, empezó a dar errores porque olvidó aplicar el mismo escalado a los clientes nuevos.

**La Solución:** Los Pipelines de Scikit-learn. No son solo un capricho de organización, sino una necesidad para evitar el **Data Leakage** (fuga de datos). El pipeline encapsula cada paso, garantizando que el tratamiento sea idéntico tanto en el entrenamiento como en la producción.

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, RobustScaler

# 1. Creamos un dataset más complejo (Numérico + Categórico)
data_banco = {
    'Monto_Compra': [100, 2500, 150, 3000, 120, 50000], # 50000 es un outlier
    'Ciudad': ['Madrid', 'Bogota', 'Madrid', 'Mexico', 'Bogota', 'Mexico']
}
df_banco = pd.DataFrame(data_banco)

print('--- Datos de Marina (Banco) ---')
print(df_banco)
print()

# 2. Definimos transformaciones específicas por tipo de columna
# RobustScaler es ideal para montos con valores atípicos (outliers)
pipeline_numerico = Pipeline(steps=[
    ('scaler', RobustScaler())
])

pipeline_categorico = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# 3. Unimos todo en un ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', pipeline_numerico, ['Monto_Compra']),
        ('cat', pipeline_categorico, ['Ciudad'])
    ]
)

# 4. Ejecución en una sola línea
resultado_final = preprocessor.fit_transform(df_banco)

print('--- Resultado del Pipeline (Matriz Limpia) ---')
print(resultado_final)
print()
print('Marina ahora tiene un proceso robusto que no se romperá en producción.')
```

### Menos es Más: Feature Selection

No todo es agregar y transformar. A veces debemos combatir la **maldición de la dimensionalidad**. Tener mil columnas no sirve de nada si solo diez explican el comportamiento. Al quedarnos solo con lo esencial, logramos:
1. Entrenamientos más rápidos.
2. Evitar el sobreajuste (overfitting).
3. Modelos más simples y elegantes.
