# Transformando Datos: Estandarización y Escalamiento Robusto

### Objetivo:
Introducir los métodos **StandardScaler** y **RobustScaler** de Scikit-Learn para ajustar la escala de los datos numéricos. Esto permite que algoritmos basados en distancias, como KNN o SVM, funcionen correctamente al tratar todas las variables con la misma importancia jerárquica.

### Introducción: El lenguaje de los datos
¡Hola! Qué bueno que sigas con nosotros porque hoy vamos a entrar en un terreno donde la matemática se vuelve pura magia para tus modelos de aprendizaje automático. Lograr que tus datos hablen el mismo idioma es la diferencia entre un modelo que adivina y uno que realmente entiende los patrones ocultos.

Imagina que intentamos comparar la altura de una hormiga en milímetros con el peso de un elefante en toneladas. Si le damos esos números brutos a un algoritmo como **KNN** o **SVM**, el peso del elefante va a dominar por completo porque sus números son gigantescos, haciendo que la altura de la hormiga sea invisible. Esto sucede porque estos modelos miden distancias espaciales. Si una variable tiene un rango de mil y otra de uno, la distancia se calculará casi exclusivamente sobre la primera, ignorando información vital de la segunda columna.

## 1. StandardScaler: El Estándar de Oro

Esta técnica realiza lo que llamamos una **transformación Z** o tipificación. Básicamente, toma cada dato, le resta el promedio de toda la columna y lo divide por la desviación estándar.

El resultado es fascinante: una distribución donde la media es exactamente cero y la dispersión (varianza) es de uno. Es ideal cuando tus datos siguen una distribución normal (la famosa campana de Gauss), porque permite que el modelo identifique qué tan lejos está un punto del centro del grupo de forma equilibrada.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# Simulación de datos de viviendas
data = {
    'habitaciones': [1, 2, 3, 4, 5, 2, 3, 4, 1, 5],
    'precio_usd': [120000, 250000, 310000, 450000, 520000, 240000, 305000, 445000, 115000, 510000]
}

df_viviendas = pd.DataFrame(data)

print('--- Datos Originales (Hormiga vs Elefante) ---')
print(df_viviendas.head())
print()

# Aplicando StandardScaler
scaler_std = StandardScaler()
viviendas_escaladas = scaler_std.fit_transform(df_viviendas)

# Convertimos a DataFrame para visualizar mejor
df_std = pd.DataFrame(viviendas_escaladas, columns=['habitaciones_std', 'precio_std'])

print('--- Datos con StandardScaler (Mismo Idioma) ---')
print(df_std.head())
print()
print('Media de las columnas:')
print(df_std.mean().round(2))
print('Desviacion estandar de las columnas:')
print(df_std.std().round(2))
```

## 2. El Peligro de los Valores Atípicos (Outliers)

Si tienes un millonario en un barrio de clase media, **StandardScaler** va a intentar promediar ese ingreso enorme y terminará aplastando a todos los demás datos hacia un rincón para compensar la balanza. La media se desplaza y la desviación estándar se infla demasiado, perdiendo la esencia de la mayoría de tu población.

Aquí es donde nos volvemos analistas estratégicos. No se trata de aplicar funciones por inercia, sino de mirar los datos y detectar ruidos que arruinen el promedio.

## 3. RobustScaler: La Solución ante el Ruido

Para solucionar este dilema, Scikit-Learn ofrece el **RobustScaler**. A diferencia de su hermano, este método no usa la media ni la desviación estándar. En su lugar, utiliza:
1. **La Mediana:** El punto medio exacto de los datos.
2. **Rango Intercuartílico (IQR):** La distancia entre el percentil 25 y el 75.

No importa si hay un gigante al final de la fila; el del medio seguirá siendo el mismo. Esto permite que el grueso de los datos mantenga su estructura a pesar de valores extremos o errores de carga.

```python
from sklearn.preprocessing import RobustScaler

# Simulacion de salarios con un error de carga (un millonario accidental)
salarios = np.array([45000, 48000, 50000, 52000, 55000, 60000, 1000000]).reshape(-1, 1)

# Aplicamos ambos metodos para comparar
std_scaler = StandardScaler()
rob_scaler = RobustScaler()

salarios_std = std_scaler.fit_transform(salarios)
salarios_rob = rob_scaler.fit_transform(salarios)

print('--- Comparativa de Escalamiento con Outliers ---')
print('Salario Original | StandardScaler | RobustScaler')
for i in range(len(salarios)):
    print(f'{salarios[i][0]} | {salarios_std[i][0]:.4f} | {salarios_rob[i][0]:.4f}')

print()
print('Nota como en RobustScaler los salarios normales estan cerca de 0,')
print('mientras que en StandardScaler todos fueron empujados a valores negativos.')
```

## Caso de Estudio: El Escenario de Lucas

Pensemos en Lucas, un analista en una empresa de logística que predice tiempos de entrega usando **KNN**. Sus datos tenían entregas fallidas que tardaban días debido a huelgas, arruinando sus promedios y haciendo que sus predicciones normales fueran pésimas.

Lucas cambió su preprocesamiento a **RobustScaler** y, mágicamente, su modelo dejó de ignorar las entregas rápidas y cotidianas. Ahora sus rutas son precisas porque el sistema ya no se distrae con esos casos extremos, logrando una eficiencia increíble en sus entregas diarias.

### ¿Cuándo elegir cada uno?
*   **StandardScaler:** Úsalo cuando tus datos estén limpios o sepas que siguen una distribución normal.
*   **RobustScaler:** Úsalo durante la exploración inicial o cuando trabajes con sensores/datos manuales que puedan tener errores o valores extremos.

Recuerda: **RobustScaler** no elimina los outliers; simplemente les quita el poder de distorsionar la escala de los demás datos.
