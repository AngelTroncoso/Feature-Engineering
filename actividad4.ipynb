## 2. El Método del Rango Intercuartílico (IQR)¿Cómo decidimos qué es demasiado grande o pequeño de forma objetiva? Necesitamos una regla matemática robusta. El **IQR** se basa en la dispersión del cincuenta por ciento central de tus datos.1. **Q1 (Percentil 25):** El valor por debajo del cual cae el 25% de los datos.2. **Q3 (Percentil 75):** El valor por debajo del cual cae el 75% de los datos.3. **IQR:** La diferencia entre Q3 y Q1.4. **Límites:** Establecemos fronteras sumando o restando 1.5 veces el IQR a los cuartiles.

```python
# 1. Calculamos los cuartilesq1 = df['precio_usd'].quantile(0.25)q3 = df['precio_usd'].quantile(0.75)# 2. Calculamos el Rango Intercuartílicoiqr = q3 - q1# 3. Definimos los límiteslimite_inferior = q1 - 1.5 * iqrlimite_superior = q3 + 1.5 * iqrprint('--- Parámetros del Filtro ---')print('Q1:', q1)print('Q3:', q3)print('IQR:', iqr)print('Límite Superior Permitido:', limite_superior)print()
```

## 3. Aplicando Capping (Truncamiento)Ahora, en lugar de eliminar la fila de la mansión (lo cual nos haría perder otros datos útiles de esa casa), usaremos la técnica de **Capping**. Transformaremos el valor extremo para que valga exactamente el límite superior. De esta manera, mantenemos el registro pero limitamos su peso exagerado en el modelo predictivo.

```python
# Aplicamos el truncamiento usando np.clipdf['precio_capping'] = df['precio_usd'].clip(lower=limite_inferior, upper=limite_superior)print('--- Comparación de Datos ---')print(df[['precio_usd', 'precio_capping']])print()# Visualización del resultado finalplt.figure(figsize=(10, 4))plt.boxplot(df['precio_capping'], vert=False)plt.title('Distribución tras aplicar Capping (IQR)')plt.xlabel('Precio USD')plt.show()print('¡Voilà! Ya tenemos un set de datos equilibrado.')
```
